# USING LABEL ENCODER TO CONVERT THE STRING TO NUMERICAL DATA

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df1['Name']=le.fit_transform(df1['Name'])

df1['Sex']=le.fit_transform(df1['Sex'])

df1['Ticket']=le.fit_transform(df1['Ticket'])
df1['Embarked']=le.fit_transform(df1['Embarked'])
df1['Cabin']=le.fit_transform(df1['Cabin'])
```

```python
df1
```

```python
df1.dtypes
```

```python
np.mean(df1['Age'])
```

```python
df1["Age"]=df1["Age"].fillna(df1["Age"].mean())
```

```python
df1['Age']=df1['Age'].astype(int)
df1['Fare']=df1['Fare'].astype(int)
df1['Cabin']=df1['Cabin'].astype(int)
```

# CHECKING FOR THE SKEWNESS OF THE DATA

```python
df1.skew()
```

```python
#Removing skewness using cubroot method

df1['SibSp']=np.cbrt(df1['SibSp'])
df1['Parch']=np.cbrt(df1['Parch'])
df1['Fare']=np.cbrt(df1['Fare'])
df1['Cabin']=np.cbrt(df1['Cabin'])
df1['Embarked']=np.cbrt(df1['Embarked'])
```

```python
df1.skew()
```

# CLEANING OF THE DATA

```python
from scipy.stats import zscore
```

```python
z=np.abs(zscore(df1))
```

```python
threshold=3
```

```python
print(np.where(z>3))
```

```python
z.iloc[863,6]
```

```python
df=df1[(z<3).all(axis=1)]

df
```

IN THIS CASE THE PASSENGERSID NEEDS TO BE REMOVED AS THERE IS NEED TO ANALYSE PASSENGERS ID

```python
df.drop('PassengerId',axis=1, inplace=True)
```

```python
x=df.iloc[:,1:11]
x
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=96)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
predlr=lr.predict(x_test)
```

```python
predlr
```

```python
accuracy_score(y_test,predlr)
```

The accuracy of the people that survived in the train using logisticRegression is 76.6%

```python
confusion_matrix(y_test,predlr)
```

```python
print(classification_report(y_test,predlr))
```
